```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

train_file_path = '/content/drive/My Drive/SML Dataset/train.csv'
test_file_path = '/content/drive/My Drive/SML Dataset/test.csv'

train_data = pd.read_csv(train_file_path)

train_description = train_data.describe()

data_types = train_data.dtypes
data_shape = train_data.shape

X_train = train_data.drop(columns=['price_range'])
y_train = train_data['price_range']

null_values = train_data.isnull().sum()
train_data_filled = train_data.fillna(train_data.mean())

scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(train_data_filled.drop(columns=['price_range']))

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train_normalized, y_train, test_size=0.2, random_state=42)

print("Train Data Description:\n", train_description)
print("\nData Types:\n", data_types)
print("\nData Shape:", data_shape)
print("\nNull Values:\n", null_values)
print("\nTraining Set Shape:", X_train_split.shape)
print("Validation Set Shape:", X_val_split.shape)
```
