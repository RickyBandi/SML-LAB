```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

train_file_path = '/content/drive/My Drive/SML Dataset/train.csv'
test_file_path = '/content/drive/My Drive/SML Dataset/test.csv'

train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

X_train = train_data.drop(columns=['price_range'])
y_train = train_data['price_range']

scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(test_data.drop(columns=['id']))

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train_normalized, y_train, test_size=0.2, random_state=42)

print("Training Set Shape:", X_train_split.shape)
print("Validation Set Shape:", X_val_split.shape)
print("Test Set Shape:", X_test_normalized.shape)
```
