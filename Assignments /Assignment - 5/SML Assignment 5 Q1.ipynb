```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from google.colab import drive

data = pd.read_csv('/content/drive/My Drive/SML Dataset/Salary_Data.csv')

X = data[['YearsExperience']]
y = data['Salary']

print(data.head())
print(X.head())
print("\nTarget:\n", y.head())

plt.scatter(X, y, color='blue', alpha=0.5)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary')
plt.show()

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
normalized_features_df = pd.DataFrame(X_normalized, columns=['YearsExperience'])
print("\nNormalized Features:\n", normalized_features_df.head())
```
